```python
df = pd.read_csv("data/Booz_Allen_Hamilton_Restricted_-_BAH1002_-_Open_Requisitions_Report_(ORR).csv", header=0)
print(df.shape)
```

```python
ml_skills = []
for idx, i in enumerate(df["Job Profile Skills"].values):
    if "machine learning" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
    elif "artificial intelligence" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
    elif "data science" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
    elif "computer vision" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
    elif "deep learning" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
    elif "large language" in str(i).replace("\n", ' ').lower():
        ml_skills.append(idx)
print(len(ml_skills))
```

```python
df.iloc[ml_skills]
```
